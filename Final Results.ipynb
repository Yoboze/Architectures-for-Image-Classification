```python
import numpy as np
import matplotlib.pyplot as plt
```

#### cifar10

```python
results1 = np.loadtxt("cifar10-resnet-relu.txt")
results2 = np.loadtxt("cifar10-resnet-Gelu.txt")
```

```python
# All have this shape
print(results1.shape)
print(results2.shape)
```

```python
plt.plot(np.arange(0,results1.shape[1]),
 np.mean(results1,0))
plt.fill_between(np.arange(0,results1.shape[1]),
 np.mean(results1,0)-(1.96*np.std(results1,0)/np.sqrt(results1.shape[0])),
 np.mean(results1,0)+(1.96*np.std(results1,0)/np.sqrt(results1.shape[0])),
 alpha=0.5)
plt.plot(np.arange(0,results2.shape[1]),
 np.mean(results2,0))
plt.fill_between(np.arange(0,results2.shape[1]),
 np.mean(results2,0)-(1.96*np.std(results2,0)/np.sqrt(results2.shape[0])),
 np.mean(results2,0)+(1.96*np.std(results2,0)/np.sqrt(results2.shape[0])),
 alpha=0.5)
#plt.ylim([0.5,1.0])
plt.ylabel('Validation Accuracy (P)')
plt.xlabel('Epoch')
plt.legend(['Cifar10-resnet-Relu','Cifar10-resnet-Gelu'])
plt.show()
```

#### Question 2
This may be happening because for me to run the architectures with gelu activation function, I need to reduce the batch size from 256 to 128. Reducing the batch size makes the architectures quick and less variability and fosters faster learning. The literature also shows that architectures using gelu activation function tend to perform better than relu activation function. For example, Dan et al. show that gelu activation function is better than relu because gelu can be more robust to noised inputs and tend to have the lowest median training log loss with and without dropout. Also, gelu are non-convex, non-monotonic activation functions and are not linear in the positive domain, which makes it exhibits curvature at all point, while relu are convex and monotonic activation function and are linear in the positive domain, thereby can lack curvature (Dan et al., 2020)



References: Dan Hendrycks and Kevin Gimpel. GAUSSIAN ERROR LINEAR UNITS (GELUS), 2020. (https://arxiv.org/pdf/1606.08415.pdf)
